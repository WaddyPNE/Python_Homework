```python
import seaborn as sns
sns.boxplot(x='Goals', y='Matches', data=EPL_df)
plt.xlabel('Goals')
plt.ylabel('Matches')
plt.title('Box Plot')
plt.show()
```

```python
club_player_goals = EPL_df.groupby(['Club', 'Name'])['Goals'].sum().reset_index()
```

```python
top_ten_goals = EPL_df.sort_values(by='Goals', ascending=False).head(11)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(top_ten_goals['Name'], top_ten_goals['Goals'], marker='o', linestyle='-')
plt.xlabel('Name')
plt.ylabel('Goals')
plt.title('Line Chart for Top Ten Goals')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better visibility
plt.show()
```

```python
top_ten_assists = EPL_df.sort_values(by='Assists', ascending=False).head(10)
```

```python
plt.figure(figsize=(20, 20))  # Set the figure size (optional)
plt.bar(top_ten_goals['Club'], top_ten_goals['Goals'], color='green')  # Create the bar chart
plt.xlabel('Club')  # Label for the x-axis
plt.ylabel('Goals')  # Label for the y-axis
plt.title('Goals Scored by Football Teams')  # Title of the chart
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Add grid lines (optional)
plt.show()  # Display the chart
```

```python

club_player_goals = EPL_df.groupby(['Club', 'Name'])['Goals'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6)) 

for player in club_player_goals['Name'].unique():
    
    player_data = club_player_goals[club_player_goals['Name'] == player]
    clubs = player_data['Club']
    goals = player_data['Goals']

    
    plt.bar(clubs, goals, label=player)

plt.xlabel('Club')  #
plt.ylabel('Total Goals Scored')  
plt.title('Total Goals Scored for Each Club by Each Player')  

plt.grid(axis='y', linestyle='--', alpha=0.7)  

plt.xticks(rotation=45) 

plt.show()  
```
